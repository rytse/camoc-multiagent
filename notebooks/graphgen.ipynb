```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N_THETA = 20
N_PHI = int(N_THETA / 2)
R = 1
```

# Simulating the manifold
We sample points and tangent vectors from the sphere.

```python
# Function whose level set yields the manifold
def g_constr(p):
    '''
    g(x, y, z) = x^2 + y^2 + z^2 - 1
    or alternatively,
    g(X) = <X, X> - 1
    '''
    return np.inner(p, p) - 1

# Gradient of g
def g_grad_constr(p):
    '''
    g'(x, y, z) = [2x, 2y, 2z]^T
    or alternatively,
    g'(X) = 2X
    '''
    return 2 * p

# Parameters for S^2
theta = np.linspace(0, 2*np.pi, N_THETA)
phi = np.linspace(0, np.pi, N_PHI)
thetav, phiv = np.meshgrid(theta, phi, indexing = 'ij')

# Generate points on the sphere
xx = R * np.cos(phiv.flatten()) * np.sin(thetav.flatten())
yy = R * np.sin(phiv.flatten()) * np.sin(thetav.flatten())
zz = R * np.cos(thetav.flatten())
pointcloud = np.rollaxis(np.array([xx, yy, zz]), 1)

# Generate tangent vectors to the sphere
uu = yy
vv = xx
ww = -2 * xx * yy / zz
vecfield = np.rollaxis(np.array([uu, vv, ww]), 1)

# Plot manifold and corresponding vector field
ax = plt.figure().add_subplot(projection='3d')
ax.scatter(xx, yy, zz)
ax.quiver(xx, yy, zz, uu, vv, ww, length=0.2, normalize=True, color='r')
plt.show()
```

# Finding the nearest triangulation

We pick a point that isn't a sampled node, and we find the simplex that it is nearest to. By the triangulation property, the simplex will be the three closest nodes.

```python
# Pick a point that isn't on the triangulation
thetar = np.pi/2
phir = 3*np.pi/2
xr = R * np.cos(phir) * np.sin(thetar)
yr = R * np.sin(phir) * np.sin(thetar)
zr = R * np.cos(thetar)
pr = np.array([xr, yr, zr])

ax = plt.figure().add_subplot(projection='3d')
ax.scatter(xx, yy, zz)
ax.scatter(xr, yr, zr, c='r')
plt.show()
```

```python
%%time

# Find closest 3 points
dvs = pointcloud - pr
dists = np.linalg.norm(dvs, axis=1)
order = np.argsort(dists)
i, j, k = order[0:3]
```

```python
ax = plt.figure().add_subplot(projection='3d')
ax.scatter(xx, yy, zz, c='k')
ax.scatter(np.array([xx[i], xx[j], xx[k]]),
           np.array([yy[i], yy[j], yy[k]]),
           np.array([zz[i], zz[j], zz[k]]),
           c='c', s=100)
ax.scatter(xr, yr, zr, c='r', s=100)
plt.show()
```

# Interpolating and projecting the action

We calculate the final action to take by
 1. Linearly interpolating the vectors attached to the nearest 3 samples
 2. Projecting onto the manifold
 3. Projecting onto the allowable action space

```python
# Interpolate
idxs = np.array([i, j, k])
vhat = np.average(vecfield[idxs], axis=0)

ax = plt.figure().add_subplot(projection='3d')
ax.scatter(xx, yy, zz, c='k')
ax.scatter(np.array([xx[i], xx[j], xx[k]]),
           np.array([yy[i], yy[j], yy[k]]),
           np.array([zz[i], zz[j], zz[k]]),
           c='c', s=100)
ax.scatter(xr, yr, zr, c='r', s=100)
ax.quiver(xx[idxs], yy[idxs], zz[idxs], uu[idxs], vv[idxs], ww[idxs], length=0.2, normalize=True, color='r')
ax.quiver(xr, yr, zr, vhat[0], vhat[1], vhat[2], length=0.2, normalize=True, color='g')
plt.show()
```

```python
# Project the vector onto the manifold using the Lagrange multiplier method
ld = 0  # calculate Lagrange multiplier via Newton's method
for i in range(2):
    dld = -g_constr(vhat + g_grad_constr(vhat) * ld) / np.inner(g_grad_constr(vhat), g_grad_constr(vhat))
    ld += dld
vbar = vhat + g_grad_constr(vhat) * ld

ax = plt.figure().add_subplot(projection='3d')
ax.scatter(xx, yy, zz, c='k')
ax.scatter(np.array([xx[i], xx[j], xx[k]]),
           np.array([yy[i], yy[j], yy[k]]),
           np.array([zz[i], zz[j], zz[k]]),
           c='c', s=100)
ax.scatter(xr, yr, zr, c='r', s=100)
ax.quiver(xx[idxs], yy[idxs], zz[idxs], uu[idxs], vv[idxs], ww[idxs], length=0.2, normalize=True, color='r')
ax.quiver(xr, yr, zr, vbar[0], vbar[1], vbar[2], length=0.2, normalize=True, color='g')
plt.show()
```
